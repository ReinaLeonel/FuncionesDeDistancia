```python
# Librerias
from scipy.spatial import distance
```

Metrica de Minkowski 

$\left( \sum_{i=1}^{n} |x_{i}-y_{i}|^{p} \right)^{\frac{1}{p}}$

para toda $p \geq 1$

```python
def minkowskiDistance(x, y, p):
    if len(x) != len(y):
        raise ValueError("Los vectores deben tener la misma longitud.")
    
    suma = 0.0
    for i in range(len(x)):
        suma += abs(x[i] - y[i]) ** p

    distancia = suma ** (1 / p)
    return distancia
```

```python

def minkowskiDistance_py(x, y, p):
    if len(x) != len(y):
        raise ValueError("Los vectores deben tener la misma longitud.")
    
    distancia = distance.minkowski(x, y, p)
    return distancia
```

Distancia coseno


```python
# main

# Vectores de ejemplo
vector1 = [1, 2, 3]
vector2 = [4, 5, 6]

# 1. Prueba con p=1 (Distancia de Manhattan)
manhattan = minkowskiDistance(vector1, vector2, 1)
manhattan_py = minkowskiDistance_py(vector1, vector2, 1)
print(f"Distancia de Manhattan entre {vector1} y {vector2}: {manhattan}")
print(f"Distancia de Manhattan (scipy) entre {vector1} y {vector2}: {manhattan_py}")

# 2. Prueba con p=2 (Distancia Euclidiana)
euclidean = minkowskiDistance(vector1, vector2, 2)
euclidean_py = minkowskiDistance_py(vector1, vector2, 2)
print(f"Distancia Euclidiana entre {vector1} y {vector2}: {euclidean}")
print(f"Distancia Euclidiana (scipy) entre {vector1} y {vector2}: {euclidean_py}")

# 3. Prueba con p=n (por ejemplo, p=3)
p = 3
distancia = minkowskiDistance(vector1, vector2, p)
distancia_py = minkowskiDistance_py(vector1, vector2, p)
print(f"Distancia de Minkowski entre {vector1} y {vector2} con p={p}: {distancia}")
print(f"Distancia de Minkowski (scipy) entre {vector1} y {vector2} con p={p}: {distancia_py}")

```
